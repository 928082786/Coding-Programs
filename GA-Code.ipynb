```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def get_generation(GM_lists):
    generation = []
    for GM in GM_lists:
        w = weight(GM)
        generation.append([GM, w])
    return generation

def random_tries(items, code_shape):
    rand_GM_lists = [np.random.randint(2, size=code_shape) for i in range(items)]
    generation = get_generation(rand_GM_lists)
    return generation

def weight(rand_GM):
    G = Matrix(GF(2),rand_GM)
    C = LinearCode(G)
    weight = C.minimum_distance()
    return weight

def choose(choices):
    weights = np.array([item[1] for item in choices])
    norm_weights = weights-np.min(weights)
    prob = norm_weights/np.sum(norm_weights)
    choice = np.random.choice(np.arange(len(choices)), p=prob)
    return choice

def breed(mum, dad):
    return np.concatenate([mum[:mum.shape[0]//2, :], dad[mum.shape[0]//2:, :]], axis=0)

def crossover(generation):
    GM_lists = []
    for i in range(0, len(generation)):
        mum = generation[choose(generation)]
        dad = generation[choose(generation)]
        GM = breed(mum[0], dad[0])
        GM_lists.append(GM)
    next_generation = get_generation(GM_lists)
    return next_generation 

def mutate(generation, epilson):
    GM_lists = []
    for item in generation:
        mutate_matrxi = np.array([np.random.choice([0,1], p=[1-epilson, epilson]) for _ in range(item[0].shape[0]*item[0].shape[1])])
        mutate_matrxi = mutate_matrxi.reshape(item[0].shape)
        mutate_item = (item[0]+mutate_matrxi)%2
        GM_lists.append(mutate_item)
    mutate_generation = get_generation(GM_lists)
    return mutate_generation 
```

```python
items = 500
epochs = 100
epilson = 0.05
code_shape = (5, 15)
generation = random_tries(items, code_shape)
max_weights_squence = []
for i in range(epochs):
    print('epoch:',i)
    generation = crossover(generation)
    generation = mutate(generation, epilson=epilson)
    weights = [item[1] for item in generation]
    max_weights_squence.append(max(weights))
```
